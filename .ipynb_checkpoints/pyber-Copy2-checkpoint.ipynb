Pyber Ride Sharing Analysis for 2016
3 observable trends:
    1. Less drivers in rural areas corresponds with higher ride fares, likely due to:
        - less competition of other ride sharing companies in the area
        - less competition from other drivers
        - distance traveled requires more gas than an urban area
    2. As you move into the urban areas, the ride fares become cheaper, likely due to:
        - more competition of other ride sharing companies
        - more competition from other drivers
        - distance traveled requires less gas than an rural area        
    3. The greatest share of total fares is found in urban areas, likely due to:
        - More drivers
        - More riders
        - Riders are less likely to own a car 
        - Less car ownership means greater usage frequency
        - Greater driver size provides greater convenience and cheaper prices 

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style="darkgrid")

# File to Load (Remember to change these)
city_data_to_load = "../Pyber/data/city_data.csv"
ride_data_to_load = "../Pyber/data/ride_data.csv"
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_ride_df = pd.merge(city_data_df, ride_data_df, how='outer', on="city")
# Display the data table for preview
city_ride_df.head()
```

```python
df1= city_ride_df
df1.rename(columns={'city':'City', 'driver_count':'Driver Count (Per City)',\
    'type':'City Type','fare':'Average Fare'}, inplace=True)
df1
```

## Bubble Plot of Ride Sharing Data

```python
df1 = city_ride_df.\
    groupby(["City Type","City"]).\
    agg({"Driver Count (Per City)":"mean",
        "Average Fare":"mean",
        "ride_id":"count"}).\
    reset_index()
df1.head()
sns.relplot(x = "ride_id",
            y = "Average Fare",
            data = df1,
            size = "Driver Count (Per City)",
            hue= "City Type",
            palette=dict(Urban="#fc5a50", Suburban="#c6fcff", \
                Rural="#dbb40c"))
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.legend(loc='upper right')
plt.savefig("../Pyber/Images/bubble_plot.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
df2 = city_ride_df.\
    groupby(["City Type"]).\
    agg({"Average Fare":"sum"}).\
    reset_index()
df2.head()

# print(sizes) # adds up to 1433, which is the total number of participants
explode = (0, 0, 0.1)  
colors = ["gold","lightskyblue","lightcoral"]

# Build Pie Chart
my_plot = df2.plot(kind='pie',y='Average Fare', ax=None, autopct='%1.1f%%',\
    startangle=140, shadow=True, labels=df2['City Type'], legend = False,\
    fontsize=12, colors=colors, explode=explode, yticks=None)
my_plot.axes.get_yaxis().set_visible(False)
#ax1.axis('equal')
plt.title("% of Total Fares by City Type")
plt.savefig("../Pyber/Images/fares_by_city_type.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
df3 = city_ride_df.\
    groupby(["City Type"]).\
    agg({"ride_id":"count"}).\
    reset_index()
df3.head()
# Build Pie Chart
myplot2 = df3.plot(kind='pie',y='ride_id', ax=None, autopct='%1.1f%%',\
    startangle=140, shadow=True, labels=df2['City Type'], legend = False,\
    fontsize=12, colors=colors, explode=explode, yticks=None)
myplot2.axes.get_yaxis().set_visible(False)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("../Pyber/Images/rides_by_city_type.png")
plt.show()
#plt.close(fig)
```

## Total Drivers by City Type

```python
df4 = city_ride_df.\
    groupby(["City Type", "City"]).\
    agg({"Driver Count (Per City)":"sum"}).\
    reset_index()
df4.head()
```

```python
df5 = df4.\
    groupby(['City Type']).\
    agg({"Driver Count (Per City)":"mean"}).\
    reset_index()
df5.head()
```

```python
# Calculate Driver Percents
df5 = df4.\
    groupby(['City Type']).\
    agg({"Driver Count (Per City)":"sum"}).\
    reset_index()
df5.head()
# Build Pie Charts
myplot3 = df5.plot(kind='pie',y='Driver Count (Per City)', ax=None, autopct='%1.1f%%',\
    startangle=140, shadow=True, labels=df5['City Type'], legend = False,\
    fontsize=12, colors=colors, explode=explode, yticks=None)
myplot3.axes.get_yaxis().set_visible(False)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Pyber/Images/driver_by_city_type.png")
plt.show()
```
